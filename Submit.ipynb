```python
training_data = pd.read_csv('../input/g-research-crypto-forecasting/train.csv')

training_data['row_id'] = training_data.index

training_data = training_data[~training_data.isin([np.nan, np.inf, -np.inf]).any(1)]

num_data = len(training_data)
training_data = training_data[:(num_data*0.5), :]

x = training_data.iloc[:,:-1].values
y = training_data.iloc[:,-1].values


imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x = imputer.fit_transform(x)

scaler = StandardScaler()
x = scaler.fit_transform(x)

my_model = XGBRegressor(objective ='reg:squarederror',n_estimators = 500)

my_model.fit(x, y)
```

```python
iter_test = env.iter_test()
for (test_df, sample_prediction_df) in iter_test:
    pred = my_model.predict(test_df)
    sample_prediction_df['Target'] = pred
    env.predict(sample_prediction_df)
```
